```python
import requests
import pandas as pd

# Fetch data from the API
response = requests.get("https://randomuser.me/api/?results=10")
data = response.json()['results']  # Extract the 'results' section

# Flattening the nested structure
flattened_data = []
for item in data:
    flattened_item = {
        'gender': item['gender'],
        'title': item['name']['title'],
        'first_name': item['name']['first'],
        'last_name': item['name']['last'],
        'street_number': item['location']['street']['number'],
        'street_name': item['location']['street']['name'],
        'city': item['location']['city'],
        'state': item['location']['state'],
        'country': item['location']['country'],
        'postcode': item['location']['postcode'],
        'latitude': item['location']['coordinates']['latitude'],
        'longitude': item['location']['coordinates']['longitude'],
        'timezone_offset': item['location']['timezone']['offset'],
        'timezone_description': item['location']['timezone']['description'],
        'email': item['email'],
        'dob_date': item['dob']['date'],
        'dob_age': item['dob']['age'],
        'phone': item['phone'],
        'nat': item['nat']
    }
    flattened_data.append(flattened_item)

# Create DataFrame
df = pd.DataFrame(flattened_data)[['gender', 'street_number', 'postcode', 'latitude', 'longitude']]

# Display DataFrame
print(df)
```

```python
import requests
import pandas as pd

url = "https://weatherapi-com.p.rapidapi.com/alerts.json"
querystring = {"q": "london"}

headers = {
    "x-rapidapi-key": "3d69894095mshfb947548e955ceep1510dfjsn518f1f00956e",
    "x-rapidapi-host": "weatherapi-com.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

if response.status_code == 200:
    try:
        data = response.json()
        
        # Inspect the structure of the response
        print("JSON Response:", data) 
        
        # If the response contains an 'alerts' section, parse it
        alerts = data.get("alerts", {}).get("alert", [])
        
        if alerts:
            df = pd.DataFrame(alerts)
            print(df)
        else:
            print("No weather alerts found for the specified location.")
    
    except ValueError:
        print("Error decoding JSON. Response content:", response.text)
else:
    print(f"Request failed with status code {response.status_code}: {response.text}")
```
